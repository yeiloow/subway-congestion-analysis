```python
import pandas as pd
from sqlalchemy import create_engine

database = "../db/subway.db"

engine = create_engine(f"sqlite:///{database}")

query = "SELECT * FROM stations JOIN station_routes ON stations.station_id = station_routes.station_id"
df = pd.read_sql(query, engine)

print(df.to_string())
```
